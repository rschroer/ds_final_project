# Going to run an average on each column in the data. 

If we can find columns where the mean is significantly greater or less than 50 that has a reasonable number of them in each category, then we can have some strong contenders.

```python
for i in range (1,82):
    field_grouping=df.columns[i]
    df_group=df[["HasDetections" ,field_grouping]].groupby(by=field_grouping).mean()
    print(df_group)
```
