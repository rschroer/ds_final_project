```python
# Doing deeper dive into DBI
# Making a function to help in the selection
def deep_dive(feature, head_number):
    ## prints out the feature
    print()
    print("==============================================")
    print(f"Feature : {feature}")
    ## Calculates nulls and prints them
    total_nulls=df[feature].isna().sum()
    print(f"Nulls: {total_nulls}")
    ## If there are nulls that exist, then it finds the mean value of Has detections for them
#     if total_nulls >0:
#         feature_nulls=df[["HasDetections",feature]].fillna(value="nullz")
#         null_group=feature_nulls[["HasDetections",feature]].groupby(by=feature).mean()
#         just_nulls= null_group.loc[null_group.index=="nullz"]
#         print()
#         print(just_nulls)    
        
    ## Getting top 10 value counts
    values_df=df[feature].value_counts()
    values_df=values_df.to_frame().head(head_number)
    print(values_df)
    ## Adding the top head_number to a list
    values_df.reset_index(inplace=True)
    top_values=values_df["index"].values
    # Make the dataframe for the signal checking
    weights_df=pd.DataFrame(columns=["HasDetections"])
    weights_df.index.name=feature
    # loop and find the mean of each value in the head_number
    for value in top_values:
        temp_df=df.loc[df[feature]==value]
        temp_df=temp_df[["HasDetections",feature]].groupby(by=feature).mean()
        weights_df=weights_df.append(temp_df)
    # Append the mean of all the null values in that column    
    if total_nulls >0:
        feature_nulls=df[["HasDetections",feature]].fillna(value="nullz")
        null_group=feature_nulls[["HasDetections",feature]].groupby(by=feature).mean()
        just_nulls= null_group.loc[null_group.index=="nullz"]
        weights_df=weights_df.append(just_nulls)         

    # Print it out    
    print()
    print(weights_df)
        
         
    

deep_dive("DefaultBrowsersIdentifier",10)
```

```python
deep_dive("AVProductStatesIdentifier",10)
```

```python
for i in range (1,82):
    deep_dive(df.columns[i],10)
```
