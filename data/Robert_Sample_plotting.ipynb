```python
## Creating Function to create some encoding for any feature

def encode_some_stuff(dataframe, column_name):    
    column_values=dataframe[column_name].unique()
    for value in column_values:
        print(value)
        dataframe[value]=0
        dataframe[value].loc[dataframe[column_name]==value]=1
        print(dataframe[value].value_counts())
    print(dataframe[column_name].value_counts())
```

```python
encode_some_stuff(df,"ProductName")
```

```python
df_group=df_columns.groupby(by="ProductName").sum()
df_group
```

```python
df_group=df_columns.groupby(by="AvSigVersion").mean()
df_group
```

```python
version_split=df_columns["AvSigVersion"].str.split(".", expand=True)
```

```python
df_columns["Subversion"]=version_split[1]
```

```python
df_columns
```

```python
version_groups=df_columns.groupby(by="Subversion").mean()
version_groups
```

```python
x_axis = np.arange(len(version_groups.index))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis)
plt.bar(x_axis, version_groups["HasDetections"])
```
